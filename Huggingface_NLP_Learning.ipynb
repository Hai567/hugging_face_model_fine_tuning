# Bias and limitations

```python
from transformers import pipeline

unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])

result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```

When asked to fill in the missing word in these two sentences, the model gives only one gender-free answer (waiter/waitress). The others are work occupations usually associated with one specific gender — and yes, prostitute ended up in the top 5 possibilities the model associates with “woman” and “work.” This happens even though BERT is one of the rare Transformer models not built by scraping data from all over the internet, but rather using apparently neutral data (it’s trained on the English Wikipedia and BookCorpus datasets).

When you use these tools, you therefore need to keep in the back of your mind that the original model you are using could very easily generate sexist, racist, or homophobic content. Fine-tuning the model on your data won’t make this intrinsic bias disappear.

# Behind the pipeline

### 1st step - Preprocessing with Tokenizer

This step is requried as we have to tokenize our data for the computers to understand

```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint) # Downloads and caches the configurations and vocabulary associated with the given checkpoint
tokenizer
```

```python
raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
] # Since the 2 input sentences do not have the same size -> We need to pad the shortest one to be able to build an array, that's why after `102` in the 2nd sentence comes with so many `0`
# This can be achieved by specifying the `padding=True`

# The type of the output tensor can also be specified
# TensorFlow tensor: tf
# PyTorch tensor: pt
# plain NumPy array: np
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt") # `truncation=True` to truncate any sentences that are longer than the model can handle

print(inputs) # The `attention_mask` tells the model the places which padding is applied so that the model do not pay attention to it.
```

### 2nd step - Model

#### AutoModel

```python
from transformers import AutoModel # AutoModel only download model bodies without their pretraining heads
# Therefore it outputs high-dimensional tensor that represents the sentences passed, which not directly useful for classification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(checkpoint) # Download and cache the configuration and pretrained weights of the model
model
```

```python
outputs = model(**inputs)
print(outputs.last_hidden_state.shape) # Represents the input sentences
# 2 sentences, 16 tokens each sentence, [hidden size]/[vector dimension of each model input] is 768
```

#### AutoModelForSequenceClassification

```python
# Get the output linked to the classification problem
from transformers import AutoModelForSequenceClassification # Works exactly like AutoModel but with classification head

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
print(outputs.logits.shape) # 2 by 2 shape for result and possibility
```

```python
print(outputs.logits) # These are not probability yet, only logits, needing to convert
```

### 3rd step - Postprocessing

```python
import torch

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

# Using with PyTorch

```python
import torch
import torch.nn.functional as F

from transformers import AutoTokenizer, AutoModelForSequenceClassification
name = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(name)
tokenizer = AutoTokenizer.from_pretrained(name)

X_train = ["I love the fact you're smiling like a genuinely happy individual.", "Good vibes. We need more that in the world."]

batch = tokenizer(X_train, padding=True, truncation=True, return_tensors="pt")
print(f'Batch: {batch} \n')

with torch.no_grad():
    outputs = model(**batch)
    print(f'Outputs: {outputs} \n')
    predictions = F.softmax(outputs.logits, 1)
    print(f"Predictions: {predictions} \n")
    labels = torch.argmax(outputs.logits, 1)
    print(f"Labels: {labels} \n")
    print(model.config.id2label)
    labels = [model.config.id2label[i] for i in labels.tolist()]
    print(labels)
```

# Finetuning

```python
# Steps:
# 1. Prepare the dataset
# 2. Load the pretrained Tokenizer -> Use it with the dataset -> Encode it
# 3. Build a PyTorch dataset with that encodings
# 4. Load pretrained models
# 5. a) Load Trainer and train your own model
#    b) Use native PyTorch training pipeline
```

```python
!wget https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz
!tar -xvzf aclImdb_v1.tar.gz
```

```python
# Create my own PyTorch dataset
import torch
import torch.nn.functional as F
import transformers
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import os

def extract_data(folder):
    texts = []
    labels = []
    for cate in ["pos", "neg"]:
        label = 1 if cate == "pos" else 0
        dir = f"./aclImdb/{folder}/{cate}"
        for filename in os.listdir(dir):
            filedir = os.path.join(dir, filename)
            with open(filedir, "r") as f:
                text = f.read()
                texts.append(text)
                labels.append(label)
    return texts, labels

class DataSet(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    def __len__(self):
        return len(self.labels)
    def __getitem__(self, index):
        item = {key: val[index].clone().detach() for key, val in self.encodings.items()} # convert every arrays within these dictionaries to PyTorch tensors.
        item["labels"] = torch.tensor(self.labels[index])
        return item
        # return self.encodings[index], self.labels[index] # Data needs to be transformed into tensor

X_train, Y_train = extract_data("train")
X_test, Y_test = extract_data("test")
```

```python
model_name = "distilbert/distilbert-base-uncased"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

train_encodings = tokenizer(X_train, padding=True, truncation=True, return_tensors="pt")
test_encodings = tokenizer(X_test, padding=True, truncation=True, return_tensors="pt")

train_dataset = DataSet(train_encodings, Y_train)
test_dataset = DataSet(test_encodings, Y_test)
```

```python
train_dataloader = DataLoader(train_dataset, batch_size = 5, shuffle = True)
test_dataloader = DataLoader(test_dataset, batch_size = 5, shuffle = True)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

optimizer = torch.optim.AdamW(model.parameters(), lr=0.001)
epochs = 5
model.to(device)
model.train()

losses = []
for oi in range(epochs):
    for i, batch in enumerate(train_dataloader):
        optimizer.zero_grad() # clear out the gradients of all Variables
        loss = model(**batch).loss
        print(f"Epoch {oi+1}/{epochs} \t batch {i}/5 \t loss: {loss}")
        losses.append(loss)
        loss.backward
        optimizer.step()
```
